```python
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
### theta = (X^T @ X)^(-1) @ X^T @ Y
```

```python
x = 5 * np.random.rand(100,1)
y= x**2-6*x+9 + np.random.randn(100,1)
plt.plot (x,y,"b.")
```

```python
X= np.concatenate((x**2,x**1,x**0),axis=1) #100*3
theta = np.linalg.inv(X.T @ X) @ X.T @ y
print(theta)
```

```python
plt.plot (x, X @ theta,"r.")
```

```python
### theta = (X^T @ X)^(-1) @ X^T @ Y
```

```python
x = 5 * np.random.rand(100,1)
y = x**3 - 9 * x**2 + 26 * x - 24 + np.random.randn(100,1)
plt.plot(x,y,"b.")
```

```python
X = np.concatenate((x**3,x**2,x**1,x**0),axis=1) #(100,4)
theta = np.linalg.inv(X.T @ X) @ X.T @ y #(4,1)
print(theta.shape)
plt.plot(x, X@theta,"r.")
print(theta)
```
